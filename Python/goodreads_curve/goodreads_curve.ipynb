```python
import pandas as pd
```

```python
df = pd.read_csv('book1-100k.csv')
df.head()
```

```python
df["RatingDist1"] = [int(x[2:]) for x in df["RatingDist1"]]
df["RatingDist2"] = [int(x[2:]) for x in df["RatingDist2"]]
df["RatingDist3"] = [int(x[2:]) for x in df["RatingDist3"]]
df["RatingDist4"] = [int(x[2:]) for x in df["RatingDist4"]]
df["RatingDist5"] = [int(x[2:]) for x in df["RatingDist5"]]
df.head()
```

```python
s1 = sum(df["RatingDist1"])
s2 = sum(df["RatingDist2"])
s3 = sum(df["RatingDist3"])
s4 = sum(df["RatingDist4"])
s5 = sum(df["RatingDist5"])
total = s1 + s2 + s3 + s4 + s5
print("1 STAR: " + str(s1/total))
print("2 STAR: " + str(s2/total))
print("3 STAR: " + str(s3/total))
print("4 STAR: " + str(s4/total))
print("5 STAR: " + str(s5/total))
```

```python
s1 * 1 / total + s2 * 2 / total + s3 * 3 / total + s4 * 4 / total + s5 * 5 / total
```
